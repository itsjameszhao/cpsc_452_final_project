```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load("./64x64_data.npz")
```

```python
input_data = data['input'][:,:,:, ::-1]
target_data = data['target']
```

```python
input_data.shape
```

```python
rows = 5
cols = 5
img_count = 0

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

for i in range(rows):
    for j in range(cols):        
        if img_count < len(input_data):
            axes[i, j].imshow(input_data[img_count])
            img_count+=1
```
