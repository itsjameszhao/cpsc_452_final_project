```python
import PIL
from PIL import Image
import glob
import os
```

```python
images = glob.glob('./faces_train/*.jpg')
file_names = [os.path.basename(path) for path in images]
```

```python
newsize = (64, 64)
for i, im_path in enumerate(images):
    img_y = Image.open(im_path)
    img_y = img_y.resize(newsize, Image.ANTIALIAS)
    img_y.save('./cropped_images/{}'.format(file_names[i]))
```
